```python
import pandas as pd
import numpy as np
import string
from typing import List
```

```python
def init_wordle():
    words = pd.read_csv('word-list.txt',header=None)
    words = words.rename(columns={0:'word'})
    words = words['word'].apply(lambda x: pd.Series(list(x)))
    words['probs'] = 1
    letters = {letter: [1 for i in range(5)] for letter in string.ascii_lowercase}
    return words, letters
words, letters = init_wordle()
```

```python
guess = "arise"
response = ['y','y','g','b','b']
def update_letters(guess: str, response: List[int]):
    """
    Takes in a guess and updates the letters and word_probability 
        guess: 5 letter word
        response: list of responses for the guessed word could be correct
        placement, correct letter, or wrong.
    """
    for position,(letter, color) in enumerate(zip(guess,response)):
        if color == 'y':
             letters[letter][position] = 0
        elif color == 'g':
            letters[letter][position] = 2
        else:
            letters[letter] = [0 for i in range(5)]
```

```python
def update_words(guess,response):
    update_letters(guess,response)
    for char in guess:
        [0, 1, 1, 1, 1]
        filters = letters[char]
        for idx, letter_status in enumerate(filters):
            if letter_status == 0:
                words['probs'] = np.where(words.iloc[:,idx] == char,0, words.probs,)
            elif letter_status == 2:
                words['probs'] = np.where(words.iloc[:,idx] != char,0, words.probs,)
update_words(guess,response)           

```

```python
def get_letter_probabilities():
    letter_probs = {}
    for char in letters:
        letter_probs[char] = [len(words[(words.iloc[:,idx] == char) & (words.probs == 1)])/
                                len(words[words.probs == 1]) 
                                for idx in range(5)]
    return letter_probs
```

```python
init_wordle()
guess= 'arise'
response = ['y', 'y', 'g', 'b', 'b']
update_words(guess,response)
get_letter_probabilities()
```
